```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
import os
```

```python
# Get current working directory
cwd = os.getcwd()
print(cwd)
```

```python
input_file ='data.csv'
input_data = pd.read_csv(input_file)
input_data
```

```python
type(input_data['Date of Contact'][0])
```

```python
input_data['Date of Contact'] = pd.to_datetime(input_data['Date of Contact'], format='%Y-%m-%d')
input_data
```

```python
type(input_data['Date of Contact'][0])
```

```python
input_data = input_data.sort_values('Date of Contact')
input_data = input_data.reset_index(drop=True)
input_data
```

```python
input_data['Month'] = input_data['Date of Contact'].apply(lambda x: x.strftime('%B')) 
```

```python
input_data
```

```python
input_data['Month'].value_counts()
```

```python
df = input_data.groupby('Month')['Client Name'].nunique()
print(df)
```

Conclusion: October has the most value counts and also the most unique client counts as well.
